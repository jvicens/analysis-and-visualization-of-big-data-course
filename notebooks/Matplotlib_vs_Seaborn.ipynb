# Plot Gallery Challenge (Matplotlib & Seaborn)

**Task:** Recreate a set of canonical plots and adapt them to your dataset.

```python
import numpy as np, pandas as pd, matplotlib.pyplot as plt, seaborn as sns
np.random.seed(0)
n = 200
df = pd.DataFrame({
    "x": np.linspace(0, 10, n),
    "y": np.sin(np.linspace(0, 10, n)) + 0.2*np.random.randn(n),
    "cat": np.random.choice(list("ABC"), size=n)
})
df.head()
```

## 1) Line plot with annotations

```python
plt.figure()
plt.plot(df["x"], df["y"])
plt.title("Line plot")
plt.xlabel("x"); plt.ylabel("y")
plt.show()
```

## 2) Histogram + KDE (Seaborn)

```python
plt.figure()
sns.histplot(df["y"], kde=True)
plt.title("Distribution of y")
plt.show()
```

## 3) Boxplot by category

```python
plt.figure()
sns.boxplot(x="cat", y="y", data=df)
plt.title("Boxplot by category")
plt.show()
```

## 4) Heatmap (correlation)

```python
plt.figure()
corr = df[["x","y"]].corr()
sns.heatmap(corr, annot=True)
plt.title("Correlation heatmap")
plt.show()
```

## 5) Pairplot (Seaborn)

```python
sns.pairplot(df[["x","y"]]); plt.show()
```
